```python
import whitebox
import glob
import rasterio
import geopandas as gpd
from rasterio.plot import show
import folium
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import json
```

```python
wbt = whitebox.WhiteboxTools()
```

```python
wbt.set_working_dir("/Users/Anthony/Data and Analysis Local/NYS_Testing/Rasters/")
```

```python
# Get all .tif files and create semicolon-separated list
tif_files = glob.glob("18*.tif")
input_string = ";".join(tif_files)

def create_nys_mosaic_if_needed(input_string):
    output_file = "NYS_combined_dem.tif"
    
    if os.path.exists(wbt.work_dir + "/" + output_file):
        print(f"{output_file} already exists. Skipping mosaic creation.")
        return False
    else:
        print(f"Creating {output_file}...")
        wbt.mosaic(
            inputs=input_string,
            output=output_file
        )
        print(f"{output_file} created successfully.")
        return True
```

```python
create_nys_mosaic_if_needed(input_string)
```
